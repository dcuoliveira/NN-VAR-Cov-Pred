```python
import pandas as pd
import os
import matplotlib.pyplot as plt

plt.style.use("bmh")
```

```python
probs = [0.05, 0.6]
p = 1
k = 300
```

```python
train = []
for ps in probs:
    tmp_train_df = pd.read_csv(os.path.join(os.getcwd(),
                                            "data",
                                            "inputs",
                                            "var_{}_{}_{}".format(ps, p, k),
                                            "betadgp_covdgp_data_train.csv"))
    
    tmp_train_df["prob"] = str(ps)
    train.append(tmp_train_df.drop(["eq", "variable"], axis=1))
train_df = pd.concat(train, axis=0)

fig, ax = plt.subplots(figsize=(8, 6))
for ps in probs:
    plt.scatter(y=train_df.loc[train_df["prob"] == str(ps)]["betas_dgp"],
                x=train_df.loc[train_df["prob"] == str(ps)]["cov_estimate"],
                label=str(ps),
                s=10,
                alpha=0.5)
plt.legend(loc='upper right')
plt.title("VAR({}) simulation with k={} and T={}".format(p, k, 100))
plt.ylabel("true phi's")
plt.xlabel("proxy phi's (covariance)")
```
